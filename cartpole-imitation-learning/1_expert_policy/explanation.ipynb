## Why a Rule-Based Controller Was Used

For this project, I used a rule-based control policy to generate expert demonstrations for the CartPole environment. The purpose of the expert policy is not to learn from trial and error, but to provide reliable state-action pairs that the imitation learning model can copy.

The CartPole state is represented as:

```text
[ x, ẋ, θ, θ̇ ]
where x is the cart position, ẋ is the cart velocity, θ is the pole angle, and θ̇  is the pole angular velocity.

A rule-based controller is appropriate for this task because CartPole is a relatively simple control problem. The main objective is to keep the pole upright by deciding whether to push the cart left or right. Since the state variables already describe the motion of the cart and pole, a simple linear feedback rule can produce strong expert actions.

The expert policy uses a weighted combination of the state variables:

````python
def expert_policy(obs):
    x, x_dot, theta, theta_dot = obs

    score = theta + 0.5 * theta_dot

    if score > 0:
        return 1   # right
    else:
        return 0   # left
````

If the score is positive, the cart is pushed to the right. If the score is negative, the cart is pushed to the left. This allows the controller to react to both the current pole angle and the direction in which the pole is moving.

I chose this method because it is simple and efficient. Compared with using a reinforcement learning algorithm to generate expert demonstrations, the rule-based controller does not require a long training process. It also makes the data collection process easier to understand because the expert action is directly determined by the current state. This is useful for imitation learning because the student policy needs a clean demonstration dataset. By using a rule-based expert, I can generate consistent state-action pairs and then train the neural network to imitate the expert’s control behavior.
